```python
sns.histplot(df["Gender"], kde=True)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x="Neighbourhood",y="No-show", data = df)
plt.show()
```

```python
# Group by neighborhood and count no-shows
counts = df.groupby("Neighbourhood")["No-show"].value_counts().unstack().fillna(0)

# Focus only on total no-shows per neighborhood
no_shows = counts["Yes"].sort_values(ascending=False)

# Plot top 10
no_shows.head(10).plot(kind="bar", figsize=(10,5), title="Top 10 Neighbourhoods by No-shows")
plt.show()
```

```python
counts = df["Neighbourhood"].value_counts()
valid_neighbourhoods = counts[counts > 500].index

filtered = df[df["Neighbourhood"].isin(valid_neighbourhoods)]
plt.figure(figsize=(15,6))
sns.countplot(x="Neighbourhood", hue="No-show", data=filtered)
plt.xticks(rotation=90)
plt.show()
```

```python
prop = (df.groupby("Neighbourhood")["No-show"]
          .value_counts(normalize=True)
          .unstack()["Yes"]
          .sort_values(ascending=False))

prop.head(10).plot(kind="bar", figsize=(10,5), title="No-show Rate by Neighbourhood (%)")
plt.show()
```
