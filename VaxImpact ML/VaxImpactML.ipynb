```python
"""
COVID-19 Outcomes by Vaccination Status – Historical Data Analysis
-------------------------------------------------------------------
Source: "COVID-19 Outcomes by Vaccination Status - Historical" dataset from City of Chicago via Data.gov (Catalog ID: 6irb-gasv). :contentReference[oaicite:0]{index=0}

Dataset Description:
- Weekly rates of COVID-19 cases, hospitalizations, and deaths among Chicago residents, stratified by age and vaccination status (Unvaccinated, Vaccinated (fully), and Boosted). :contentReference[oaicite:1]{index=1}
- Includes demographic age groupings and population counts; excludes partially vaccinated individuals. :contentReference[oaicite:2]{index=2}

What We Did:
1. Loaded and cleaned the data: parsed dates, handled missing values, filtered out rows with no outcomes, created numeric features (e.g., week index, vaccination/booster rates, age group ranges).  
2. Built regression models (DecisionTreeRegressor and RandomForestRegressor) to predict weekly outcome counts (cases/hospitalizations/deaths) separately for each vaccination status (Unvaccinated, Vaccinated, Boosted).  
3. Compared model performance with metrics MSE and R² to assess which model and vaccination status grouping yield the most accurate predictions.  
4. Visualized results to highlight differences in predicted vs actual outcomes by vaccine status, illustrating the effectiveness of vaccination and booster doses.

Author: Marwan Mohamed Sayed, MBA
"""
```

```python

import pandas as pd
import sklearn
import numpy
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

```

```python
df = pd.read_csv("COVID_Data_2022.csv")
print(df.head())
print(df.info())
print(df.isnull().sum())
```

```python
df = df.dropna(subset=["Outcome Unvaccinated", "Outcome Vaccinated", "Outcome Boosted"]) 
```

```python
#filling no data with medians for numeric values only
df = df.fillna(df.median(numeric_only=True))
```

```python
#for rows that have 0 in outcomes or population (likely noise, reporting delay, or missing data).
#dropping rows that have all three outcomes = 0 (not useful for training).
#dropping rows with population = 0 
df = df[(df["Outcome Unvaccinated"] + df["Outcome Vaccinated"] + df["Outcome Boosted"]) > 0]
df = df[(df["Population Unvaccinated"] > 0) & 
        (df["Population Vaccinated"] > 0) & 
        (df["Population Boosted"] > 0)]
df = df.dropna()
```

```python
#turning "Week End" dates into week index (0 = first week, increasing by 1 each week) 
df["Week End"] = pd.to_datetime(df["Week End"])
df["Week_Num"] = (df["Week End"] - df["Week End"].min()).dt.days // 7
weeks = df["Week_Num"]

```

```python
print(df.describe())

```

```python
"""
Decision Tree Regression Model for Predicting Weekly Unvaccinated COVID-19 Outcomes

Process:
1. Selected features based on relevance to predicting weekly unvaccinated outcomes:
   - Week_Num (numeric week index)
   - Age Group Min & Max
   - Vaccinated Rate, Boosted Rate
   - Population Unvaccinated, Population Vaccinated, Population Boosted
2. Split dataset into training (80%) and testing (20%) sets to evaluate model performance on unseen data.
3. Trained a DecisionTreeRegressor with a fixed random state and depth to capture patterns in the data without overfitting.
4. Evaluated model using regression metrics (Mean Squared Error and R²) to measure prediction accuracy and variance explained.
5. Decision Trees were chosen as a starting point due to simplicity, interpretability, and their ability to handle non-linear relationships in tabular data.
"""


X_unvax= df[["Week_Num", "Age Group Min", "Age Group Max", 
        "Vaccinated Rate", "Boosted Rate", 
        "Population Unvaccinated", "Population Vaccinated", "Population Boosted"]]

y_unvax = df["Outcome Unvaccinated"]

X_train_unvax, X_test_unvax, y_train_unvax, y_test_unvax = train_test_split(X_unvax, y_unvax, test_size=0.2, random_state=42)
```

```python
model = DecisionTreeRegressor(random_state=42, max_depth=5)
model.fit(X_train_unvax, y_train_unvax)
```

```python
y_pred = model.predict(X_test_unvax)
print("MSE:", mean_squared_error(y_test_unvax, y_pred))
print("R²:", r2_score(y_test_unvax, y_pred))
```

```python

"""
Model Results Interpretation

- Mean Squared Error (MSE): Measures the average squared difference between predicted and actual outcomes.
  Lower values indicate predictions are closer to actual values.

- R² (Coefficient of Determination): Measures how much of the variance in the actual outcomes is explained by the model.
  R² = 0.72 indicates that 72% of the variation in weekly unvaccinated outcomes is captured by the model.

Takeaways:
- The Decision Tree captures most of the signal in the data and provides a solid baseline for predicting weekly unvaccinated outcomes.
- Further improvements can be achieved using ensemble methods like Random Forests, hyperparameter tuning, or additional feature engineering.

Mean Squared Error (MSE) Interpretation:

- MSE = 116,274 indicates the average squared difference between predicted and actual weekly unvaccinated counts.
- Taking the square root gives an approximate average error of ±341 unvaccinated cases per week.
- Given the variability in weekly case numbers, this shows the Decision Tree is capturing the general trends reasonably well.
- MSE serves as a baseline; using Random Forests or additional features could reduce this error further.
"""
```

```python
X_vax = df[["Week_Num", "Age Group Min", "Age Group Max", 
        "Vaccinated Rate", "Boosted Rate", 
        "Population Unvaccinated", "Population Vaccinated", "Population Boosted"]]

y_vax = df["Outcome Vaccinated"]

X_train_vax, X_test_vax, y_train_vax, y_test_vax = train_test_split(X_vax, y_vax, test_size=0.2, random_state=42)

model = DecisionTreeRegressor(random_state=42, max_depth=5)
model.fit(X_train_vax, y_train_vax)

y_pred2 = model.predict(X_test_vax)
print("MSE:", mean_squared_error(y_test_vax, y_pred2))
print("R²:", r2_score(y_test_vax, y_pred2))
```

```python
"""
DecisionTreeRegressor Results (Vaccinated Outcomes)

The DecisionTreeRegressor was trained to predict weekly COVID-19 outcomes 
(hospitalizations or deaths) among vaccinated individuals, using features 
like week index, age group, vaccination/booster rates, and population sizes.

Results:
- Mean Squared Error (MSE): 48,204 → On average, predictions were off by about
  219 cases per week (sqrt(MSE)).
- R² Score: 0.816 → The model explains ~82% of the variation in vaccinated outcomes.

Interpretation:
The decision tree captured strong relationships between vaccination status, 
population dynamics, and weekly outcomes. While reasonably accurate, it still 
left ~18% of variance unexplained, showing some limits in model depth/splitting.
"""
```

```python
X_vax = df[["Week_Num", "Age Group Min", "Age Group Max", 
        "Vaccinated Rate", "Boosted Rate", 
        "Population Unvaccinated", "Population Vaccinated", "Population Boosted"]]

y_vax = df["Outcome Vaccinated"]

X_train_vax, X_test_vax, y_train_vax, y_test_vax = train_test_split(X_vax, y_vax, test_size=0.2, random_state=42)

model = RandomForestRegressor(n_estimators=100 ,random_state=42, max_depth=5)
model.fit(X_train_vax, y_train_vax)

y_pred3 = model.predict(X_test_vax)
print("MSE:", mean_squared_error(y_test_vax, y_pred3))
print("R²:", r2_score(y_test_vax, y_pred3))
```

```python
"""
RandomForestRegressor Results (Vaccinated Outcomes)

The RandomForestRegressor (an ensemble of decision trees) was trained on the 
same vaccinated outcomes dataset as the single decision tree, using week index, 
age group, vaccination/booster rates, and population sizes as predictors.

Results:
- Mean Squared Error (MSE): 29,248 → On average, predictions were off by about
  171 cases per week (sqrt(MSE)).
- R² Score: 0.888 → The model explains ~89% of the variation in vaccinated outcomes.

Interpretation:
Compared to the single decision tree, the random forest significantly reduced 
prediction error and explained more variance. This shows that ensemble learning 
is more effective at capturing complex interactions and reducing overfitting. 
The forest provides a stronger basis for comparing vaccinated vs unvaccinated outcomes.
"""
```

```python
X_boost = df[["Week_Num", "Age Group Min", "Age Group Max", 
        "Vaccinated Rate", "Boosted Rate", 
        "Population Unvaccinated", "Population Vaccinated", "Population Boosted"]]

y_boost = df["Outcome Boosted"]

X_train_boost, X_test_boost, y_train_boost, y_test_boost = train_test_split(X_boost, y_boost, test_size=0.2, random_state=42)

model = DecisionTreeRegressor(random_state=42, max_depth=5)
model.fit(X_train_boost, y_train_boost)

y_pred4 = model.predict(X_test_boost)
print("MSE:", mean_squared_error(y_test_boost, y_pred4))
print("R²:", r2_score(y_test_boost, y_pred4))
```

```python
"""
DecisionTreeRegressor Results (Boosted Outcome):
------------------------------------------------
- Mean Squared Error (MSE): 11505.14
- R² Score: 0.9473

Interpretation:
The DecisionTreeRegressor explains approximately 95% of the variance in the boosted outcome variable, 
with moderate prediction error. While the performance is strong, Decision Trees can overfit and may not 
generalize as well compared to ensemble methods.
"""
```

```python
X_boost = df[["Week_Num", "Age Group Min", "Age Group Max", 
        "Vaccinated Rate", "Boosted Rate", 
        "Population Unvaccinated", "Population Vaccinated", "Population Boosted"]]

y_boost = df["Outcome Boosted"]

X_train_boost, X_test_boost, y_train_boost, y_test_boost = train_test_split(X_boost, y_boost, test_size=0.2, random_state=42)

model = RandomForestRegressor(n_estimators= 100, random_state=42, max_depth=5)
model.fit(X_train_boost, y_train_boost)

y_pred5 = model.predict(X_test_boost)
print("MSE:", mean_squared_error(y_test_boost, y_pred5))
print("R²:", r2_score(y_test_boost, y_pred5))
```
